#### `merchant`


```python
cols = ["merchant", "category", 'merch_lat','merch_long']
check = df[cols]
check["merch_coord"] = check["merch_lat"].astype(str) + ", " + check["merch_long"].astype(str)
display(check.head())

cols += ['merch_coord']
for col in cols:
    n_values = check[col].nunique()
    print(f"{col}: {n_values}")
```

merchant coordinates are unique for each record and doesn't make much sense.
For this application, will remove `merch_lat` and `merch_long` instead

```python
# removing merch_lat and merch_long
cols = ["merch_lat", "merch_long"]
df = df.drop(cols, axis="columns")

cols_left = [x for x in cols_left if not x in cols or cols.remove(x)]
print(f"Columns remaining: {cols_left}")
```

```python
cols = ["merchant", "category"]
merc = df[cols].drop_duplicates().sort_values("merchant").reset_index(drop=True)
print(merc.shape)
display(merc.head(20))
for col in cols:
    n_values = check[col].nunique()
    print(f"{col}: {n_values}")

cols_left = [x for x in cols_left if not x in cols or cols.remove(x)]
print(f"Columns remaining: {cols_left}")
```

```python
merc[merc.duplicated(subset=['merchant'], keep=False)].sort_values('merchant').head()
```

Definitely plausible for merchants to have transactions of different category.   
Given we are going for 3rd Normal Form (OLTP), merchant and category should be separate tables

```python
cat = merc[['category']].drop_duplicates()
print(cat.shape)
display(cat.head())

merc = merc[['merchant']].drop_duplicates()
print(merc.shape)
display(merc.head())
```
